```python
import pandas as pd
df = pd.read_csv("books.csv")
df.head(10)
```

```python
df = df.dropna()
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.utils import embedding_functions
```

```python
df["search_text"] = (
    "Title: " + df["title"] + "| Author: " + df["authors"] + "| Categories: " + df["categories"] + "| Summary: " + df["description"]
)
```

```python
df.head(10)
```

```python
df["search_text"]
```

```python
model = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")
client = chromadb.PersistentClient(path="./book_vector_db")
collection = client.get_or_create_collection(name="new_book_catalog", embedding_function=model)
```

```python
collection.count()
```

```python
collection
```

```python
p = df["search_text"].tolist()
p
```

```python
if collection.count() == 0:
    print("Embedding getting ready...")
    
    collection.add(
        ids=[str(i) for i in df.index],
        
        documents=df["search_text"].tolist(),
        metadatas=df[['title', 'authors', 'average_rating']].to_dict(orient='records')
    )
    print("Database = ready")
```

```python
collection.count()
```

```python
vs=df[['title', 'authors', 'average_rating']].to_dict(orient='records')
vs
```

```python
def recommend_books(user_query, num_results=5):
    results = collection.query(
        query_texts = [user_query],
        n_results = num_results
        
    )
    print(f"\nTop {num_results} for: '{user_query}'")
    for i in range(num_results):
        res = results['metadatas'][0][i]
        score = results['distances'][0][i]
        print(f"- {res['title']} by {res['authors']} (Rating: {res['average_rating']}) [Dist: {round(score, 2)}]")
```

```python
recommend_books("Detective and mystery stories")
```
